%Ecological and evolutionary processes are mixed together to drive population density and trait changes 

%Most intuitive examples are biotic or abiotic factors driving trait and density changes within populations.

%----------------------------------------------------------------------------------------------------------------

%One of the most fundamental theorems in evolution to understand trait change is the Price equation 

%---------------------------------------------------------------------------------------------------------------------

%The Price equation has two terms, the first describing the mean phenotypic change and the second transmission bias -- In ecological terms, selecting factors can be many, abiotic like climate or habitat conditions, and biotic like predators, parasites or others. The transmission bias change can be due to change in the environment, mutations, or other forces not explicitly introduced in the equation

%---------------------------------------------------------------------------------------------------------------------

%It can be represented (using LateX symbols) as

$$\Delta \overline{\phi} = \frac{1}{\overline{w}} (cov(w,\phi) + E(w \delta))$$

%LateX example$$e^x=\sum_{i=0}^\infty \frac{1}{i!}x^i$$

%Further reading
                                     
%Gardner, A. (2008). The Price Equation. Current Biology, 18:PR198-R202.
                                     
%Queller, D. C. (2017). Fundamental Theorems of Evolution. The American Naturalist, 189:345-353.


```octave
%A simple example in octave of fitness dependent (w) of a trait value for one resource and one consumer, 
%$\phi$, can be expressed as (Check octave free-software at $https://www.gnu.org/software/octave/$)

ecoevo1 %plots Figure 1
```

%... the resource fitness function producing the top right hand side plot above is 

$$w(\phi) = \frac{1}{1 + e^{-\gamma(\phi_{R} - \overline{\phi}_{C})^2}}$$

```octave
%which in octave is written as

%Fitness function for resources==================================================================

gamma = 5;%Strength of interaction between resource (trait value Zr) and consumer (trait value Zc)
for pB = 1:length(Zr);
    WB(pB,2) = 1/(1 + exp(-gamma*(Zr(1,pB) - mean(Zc))^2));
    WB(pB,1) = Zr(1,pB) - mean(Zc);
end
```

%and the consumer fitness function producing the bottom right hand side plot above is 

$$w(\phi) = e^{-\gamma(\phi_{R} - \overline{\phi}_{C})^2}$$
