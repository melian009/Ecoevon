### Feedbacks between coevolving predator-prey interactions and the functioning of aquatic ecosystems


### ---------------------------Intro--------------------------------------------------------------------------------
Background and rationale – 
There is a consensus among scientists that eutrophication and global climatic changes will likely promote more intense and long-lasting cyanobacterial blooms. These blooms are considered bottlenecks for
carbon transfer along the food chains, supporting lower secondary production and biodiversity and acting as a driving force in aquatic communities and on ecosystem services. Despite the fact that some of the effects of cyanobacterial
blooms are well understood, bloom-regulating mechanisms are less well known. In this regard, the role of reciprocal changes between defense and tolerance traits in cyanobacteria-zooplankton interactions and their effect on ecosystem
functioning determining the structure and the stability of cyanobacterial blooms is still at a very incipient stage. It is known that predator-prey interaction is a key factor in community structuring, with size, shape, attack and inducible
defense traits being primary aspects in determining the structure of plankton communities. Among herbivorous zooplankton, Daphnia is considered the main zooplankton preying on phytoplankton, being largely affected by the
composition, stoichiometry and biochemistry of its prey. Several cyanobacteria genera and strains are toxin-producers, able to cause several deleterious effects in zooplankton, including decreased mobility, survivorship and fecundity. Low nutritional value and morphology of cyanobacteria are also aspects that constrain zooplankton fitness. One of the
biggest issues in cyanobacteria-zooplankton interactions is how important defense traits are for the rapid growth of toxic cyanobacteria. Concerning the predation aspect, zooplankton tolerance is also considered an important trait (i.e., counter-defense) as it may determine zooplankton composition and control cyanobacterial blooms. Both toxin-producer defenses and predator tolerance display many variants, and how genotype and phenotype variation in cyanobacteria
(toxic/non-toxic) and Daphnia (sensitive/resistant) influences their interaction and their coevolutionary rate change for predicting blooms is not well understood in a rapidly changing world.
Aims – Our overall goal is to understand the feedbacks between species interactions and ecosystem functioning in aquatic ecosystems. To achieve this, we will combine models and experiments to explore how coevolving predator-
prey interactions affect ecosystem states (blooming), and how the ecosystem changes the feedback concerning the dynamics of species interactions. We aim to understand: (i) the effects of bloom-forming cyanobacteria on predator
feeding traits, behavior and population dynamics; (ii) the biological interactions underlying cyanobacterial blooms: cyanobacterial genotype and population dynamics, predator trait selection and induced defenses; (iii) the feedbacks between co-evolutionary interactions and ecosystem dynamics and (iv) ecosystem state transitions and resilience in evolving multispecies communities.
Methods – We propose microcosm laboratory and outdoor mesocosm experiments in order to test the effects of bloom-forming cyanobacteria on predator feeding behavior and its population dynamics. We will perform a second experiment
to test cyanobacterial genotype dynamics, its effects on predator traits selection and induced defenses in both organisms. We will examine how toxins might drive the interaction strength between cyanobacterial and predator
strains and how nutrient input might be influencing these interactions and the probability to produce blooms. We will build models to connect the experimental data obtained concerning defense and tolerance traits with the feedbacks of
species interactions and ecosystem functioning. We will also extend existing models considering reciprocal defense-tolerance trait changes by adding ecosystem functioning variables in order to study the feedbacks of predator-prey
coevolution, blooming and ecosystem state
